```python
from sklearn.datasets import load_boston
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

提取后6列作为数据

```python
def dealdata():
    boston=load_boston()
    dealeddata=[]
    for i in boston.data:
        dealeddata+=[i[7:]]
    return dealeddata
```

```python
def selfpca(data,window=10):#window为窗口大小
    dealeddata=data
    step=506//window#设定窗口个数
    count=0#记录步骤
    pca=PCA(n_components=2)
    
    fig, ax = plt.subplots()
    plt.xlim((0, step+1))
    plt.ylim()
    plt.xlabel('step')
    plt.ylabel('rate')
    while count<step:
        windows=dealeddata[count*window:(count+1)*window]#每个时间窗口内的数据,pca12为第一，第二主成分
        if count==0:
            pca12=pca.fit_transform(windows)
            ev12=pca.fit(windows).explained_variance_ratio_#第一、二主成分信息占比
        else:
            pca12=np.vstack((pca12,pca.fit_transform(windows)))
            ev12=np.vstack((ev12,pca.fit(windows).explained_variance_ratio_))#第一、二主成分信息占比
        count+=1
    ev1=[i[0] for i in ev12] 
    ev1plus2=[i[1]+i[0] for i in ev12]
    ax.plot([i for i in range(1,step+1)],ev1,'ro-',label='pca1')
    ax.plot([i for i in range(1,step+1)],ev1plus2,'bo-',label='pca1+pca2')
    pca1=[i[0] for i in pca12]
    pca2=[i[1] for i in pca12]#此为所求第一、第二主成分
    ax.legend()
    count2=0
    '''
    分步打印第一、第二主成分
    while count2<step:
        print('step:',count2+1)
        print('pca1:',np.array(pca1[count2*window:(count2+1)*window]))
        print('pca2:',np.array(pca2[count2*window:(count2+1)*window]))
        count2+=1
    '''
```

```python
if __name__=="__main__":
    dealeddata=dealdata()
    selfpca(dealeddata)
```
